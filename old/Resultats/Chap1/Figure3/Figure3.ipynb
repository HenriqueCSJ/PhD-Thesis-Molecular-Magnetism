```python
#On charge les librairies
import matplotlib.gridspec as gridspec
from matplotlib.patches import Ellipse
import os
#locale
sys.path.append("/home/hukadan/These/")
from setproc.cycle_process.classes.cycle_process import CycleProcess
from setproc.sweep_set.functions.filter import filter
execfile("/home/hukadan/These/Manuscript/Theorie/MagMol/scripts/TbPc2Nuclear.py")

#On prepare les figures
#premiere grid pour step et pics
gs1 = gridspec.GridSpec(2,1)
gs1.update(left=0.08, right=0.45,hspace=0,bottom=0.55, top=0.98)
ax1 = plt.subplot(gs1[0,0])
ax2 = plt.subplot(gs1[1,0])
#Pour la stat des sauts
gs2 = gridspec.GridSpec(2,1)
gs2.update(left=0.08, right=0.45,hspace=0.1,bottom=0.1, top=0.4)
ax3 = plt.subplot(gs2[:,:])
#Pour les resonances
gs3 = gridspec.GridSpec(2,1)
gs3.update(left=0.55, right=0.98,hspace=0,bottom=0.1, top=0.98)
ax4 = plt.subplot(gs3[0,0])
ax5 = plt.subplot(gs3[1,0])

fig = gcf()
fig.set_size_inches(12,7)


#On charge les donnee
os.chdir("/media/Iomega_HDD/Measures/Tb2C/sample3/6R/J9/Cycle/Cycle16/")
Cyc16 = CycleProcess("G_B_trace.json","G_B_retrace.json",[0,1,2,3,4,5,6],"Json")
Cyc16.GetStat(150,4)


#panel a et b
#coube exemple
p1 = Cyc16.trace["data"][17]
p2 = Cyc16.trace["data"][76]
bias = Cyc16.trace['bias']
p1f = filter(array(p1),4,1,1)
p2f = filter(array(p2),4,1,1)
biasf = bias[6:]

ax1.plot(bias, p1)
ax1.plot(bias, p2)
ax2.plot(biasf, p1f)
ax2.plot(biasf, p2f)
ax1.set_xlim(-0.1,0.1)
ax1.set_ylim(1.88e-6,1.91e-6)
ax2.set_xlim(-0.1,0.1)
ax2.set_ylim(-2e-29,1e-29)

#panel c
#stat des sauts
statvalue = []
for x in Cyc16['detection'] :
	statvalue.append(log10(abs(x.value)))
tr0 = Polygon([[-29.5,0],[-31.5,0],[-31.5,700],[-29.5,700]], color="#C0C0C0")
ax3.add_patch(tr0)
ax3.hist(statvalue,200)
ax3.plot([-29.5,-29.5],[0,700],'--',color='gray')

#panel d
#Zeeman
R = Tb_Pc2_ZeemanNuclear(-0.1,0.1,250,0,0)
B = linspace(-0.1,0.1,250)
ax4.plot(B,R,color="red")
ax4.set_xlim(-0.05,0.05)
ax4.set_ylim(-645.1,-644)


r1 = Ellipse((-0.037,-644.55),0.01,0.01/0.1,facecolor="#C0C0C0")
ax4.add_patch(r1)
r1 = Ellipse((-0.0125,-644.58),0.01,0.01/0.1,facecolor="#C0C0C0")
ax4.add_patch(r1)
r1 = Ellipse((0.037,-644.55),0.01,0.01/0.1,facecolor="#C0C0C0")
ax4.add_patch(r1)
r1 = Ellipse((0.0125,-644.58),0.01,0.01/0.1,facecolor="#C0C0C0")
ax4.add_patch(r1)

#panel e
stat_tr = []
for x in Cyc16["detection"] :
    if x.trace :
        stat_tr.append(x.field-0.0234) #on compense l'offset lockin

ax5.hist(stat_tr,700)
ax5.set_xlim(-0.05,0.05)



#On mets tout ca en page
#panel 1
ax1.yaxis.set_ticks([1.885e-6,1.895e-6,1.905e-6])
ax1.yaxis.set_ticklabels([1.885,1.895,1.905])
ax1.xaxis.set_visible(False)
ax1.set_ylabel(r"d$I$/d$V$ ($\mu$S)")

ax2.yaxis.set_ticks([-1.5e-29,-1.0e-29,-0.5e-29,0.0e-29,0.5e-29])
ax2.yaxis.set_ticklabels([-1.5,-1.0,-0.5,0.0,0.5])
ax2.xaxis.set_ticks([-0.05,0.05])
ax2.xaxis.set_ticklabels([-50,50])
ax2.set_ylabel(r"u.a.")
ax2.set_xlabel(r"$B$ (mT)")

#panel 4
ax4.yaxis.set_ticks([-645,-644.6,-644.2])
ax4.xaxis.set_visible(False)
ax4.set_ylabel(r"$E$ (K)")

ax5.yaxis.set_ticks([50,100,150])
ax5.set_ylabel(r"$Coups$")
ax5.xaxis.set_ticks([-0.04,-0.02,0,0.02,0.04])
ax5.xaxis.set_ticklabels([-40,-20,0,20,40])
ax5.set_xlabel(r"$B$ (mT)")

#panel 3
ax3.yaxis.set_ticks([200,400,600])
ax3.set_ylabel(r"$Coups$")
ax3.xaxis.set_ticks([-31,-30,-29,-28])
ax3.set_xlabel(r"hauteur de saut (u.a.)")
fig.savefig("/home/hukadan/These/Manuscript/Resultats/Chap1/Figure3/figure3.pdf")
```
